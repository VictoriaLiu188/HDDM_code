```python
import csv
import os

# List of input CSV files
target_folder =  '/data/victoria/HDDM_code'
os.chdir(target_folder)
# input_csv_files = ['roi1_mode.csv', 'roi2_mode.csv', 'roi8_mode.csv']  # Add your file names here
roi_ls = [1,2,8,14,15,16,17,18,20,23,24,25,28,29,31,32,33,34,35,38,39,40,41,42,43,44,48,51,52,53,60,61,65,67,68,70,74,77,\
    78,79,80,81,82,83,88,90,91,92,93,94,99,105,106]

# Output CSV file
output_csv_file = 'combined_output.csv'

# Function to extract content from input CSV and write to output CSV
def process_csv(input_file, output_csv_writer):
    with open(input_file, 'r') as input_csv:
        csv_reader = csv.reader(input_csv)
        for row in csv_reader:
            output_csv_writer.writerow(row)

# Open the output CSV file
with open(output_csv_file, 'w', newline='') as output_csv:
    csv_writer = csv.writer(output_csv)

    # Process each input CSV file
    for i in range(len(roi_ls)):
        input_file = f"roi{roi_ls[i]}_mode.csv"
        if os.path.exists(input_file):
            process_csv(input_file, csv_writer)
            print(f"Processed: {input_file}")
        else:
            print(f"File not found: {input_file}")

print("All files processed and combined into", output_csv_file)
```

```python
import csv
import os

# List of input CSV files
target_folder = '/data/victoria/HDDM_code'
os.chdir(target_folder)
roi_ls = [38,39,40,41,42,43,44,48,90,91,92,93,94] # only DMN regions this time

# Output CSV file
output_csv_file = 'combined_actual_value_mode.csv'

# Function to extract content from input CSV and write to output CSV without rows containing 'node_name'
def process_csv_without_node_name(input_file, csv_writer):
    with open(input_file, 'r') as input_csv:
        csv_reader = csv.reader(input_csv)
        for row in csv_reader:
            if 'node_name' not in row:  # Check if 'node_name' is not present in any cell of the row
                csv_writer.writerow(row)

# Open the output CSV file
with open(output_csv_file, 'w', newline='') as output_csv:
    csv_writer = csv.writer(output_csv)
    
    # Process each input CSV file
    for i in range(len(roi_ls)):
        input_file = f"roi{roi_ls[i]}_mode.csv"
        if os.path.exists(input_file):
            process_csv_without_node_name(input_file, csv_writer)
            print(f"Processed: {input_file}")
        else:
            print(f"File not found: {input_file}")

print("All files processed and combined without rows containing 'node_name' into", output_csv_file)
```

```python
import pandas as pd

# Input CSV file path
input_csv_file = 'posteriors_w_para_roi.csv'  # Replace with the actual file path

# Read CSV into a DataFrame
df = pd.read_csv(input_csv_file)

DMN_ls = [39,40,41,44,92,38,39,42,43,48,90,91,93,94]
DA_ls = [16,17,67,14,15,18,20,65,67,68,70]
control_ls = [31,32,33,80,82,83,88,34,35]
salience = [74,23,24,25,77]
viscent = [51,52,53,1,2]
# Drop the first column
df['network'] = '' 
for index, row in df.iterrows():
    if row['roi'] in DMN_ls:
        df.at[index, 'network'] = 'DMN'
    elif row['roi'] in DA_ls:
        df.at[index, 'network'] = 'DorsalAttention'
    elif row['roi'] in salience:
        df.at[index, 'network'] = 'salience'
    elif row['roi'] in control_ls:
        df.at[index, 'network'] = 'control'
    elif row['roi'] in viscent:
        df.at[index, 'network'] = 'VisualCentral'
    else:
        df.at[index, 'network'] = 'limbic'
df.to_csv('mode_w_network.csv')

```

```python
import pandas as pd

# Load the CSV file into a DataFrame
csv_file_path = 'mode_w_network.csv'
df = pd.read_csv(csv_file_path)

# Initialize an empty list to store the 'stimuli type' values
stimuli_type_values = []

# Loop through the rows and populate 'stimuli type' based on 'node' column
for index, row in df.iterrows():
    node = row['node_name']
    if 'Intercept' in node:
        stimuli_type_values.append('prototype')
    elif 'exception' in node and 'prototype' in node:
        if 'roi' in node:
            stimuli_type_values.append('exception_roi')
        else:
            stimuli_type_values.append('exception')
    elif 'rule follower' in node and 'prototype' in node:
        if 'roi' in node:
            stimuli_type_values.append('rule follower_roi')
        else: 
            stimuli_type_values.append('rule follower')
    else:
        stimuli_type_values.append('prototype_roi')

# Add the 'stimuli type' column to the DataFrame
df['stimuli type'] = stimuli_type_values

# Save the updated DataFrame to a new CSV file

df.to_csv('mode_full.csv')
```
