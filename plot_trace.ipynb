```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(0)

# Generate random data
num_rows = 3000
num_columns = 6
data = np.random.rand(num_rows, num_columns)

# Plot the data by category
fig, ax = plt.subplots()

# Assign each column to a different color
colors = ['red', 'green', 'blue', 'purple', 'orange', 'pink']

for i in range(num_columns):
    ax.scatter(data[:, i], np.arange(num_rows), label=f'Column {i + 1}', c=colors[i])

ax.set_xlabel('Value')
ax.set_ylabel('Row Index')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(0)

# Generate random data
num_rows = 3000
num_columns = 6
data = np.random.rand(num_rows, num_columns)

# Create subplots for each column
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
fig.suptitle('Histograms for Each Column', fontsize=16)

for i in range(num_columns):
    row, col = i // 3, i % 3
    axes[row, col].hist(data[:, i], bins=20, color='skyblue', edgecolor='black')
    axes[row, col].set_title(f'Column {i + 1}')
    axes[row, col].set_xlabel('Value')
    axes[row, col].set_ylabel('Frequency')

# Adjust layout for better readability
plt.tight_layout()
plt.subplots_adjust(top=0.9)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your data
data = [
    [-0.369099146, -2.079008834, 0.825780916, 1.709909688, -2.90478975, -1.194880061],
    [1.092418074, -1.210577718, 0.995315257, 2.302995792, -2.205892974, 0.097102818],
    [0.642938628, -1.389623938, 0.042496798, 2.032562566, -1.432120736, 0.600441829],
    [0.55119586, -0.621548761, -0.273297145, 1.172744621, -0.348251615, 0.824493006],
    [-0.033951468, -1.023391438, 0.774718211, 0.98943997, -1.798109648, -0.808669679],
    [0.422721246, -1.078274681, 0.282137389, 1.500995928, -1.36041207, 0.140583858],
]

# Transpose the data
data_transposed = np.array(data).T

# Column names
column_names = ["proto_slopes", "rulefollower_slopes", "exception_slopes", "proto_rf", "rf_excep", "proto_excep"]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot each column as separate points without connecting them
for i, column_data in enumerate(data_transposed):
    ax.scatter([i] * len(column_data), column_data, marker='o', label=column_names[i])

# Set x-axis labels
ax.set_xticks(range(len(column_names)))
ax.set_xticklabels(column_names)
plt.xticks(rotation=45)

# Set labels and title
plt.xlabel('Column Names')
plt.ylabel('Values')
plt.title('Data Points')

# Show a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
