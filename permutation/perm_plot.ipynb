```python
import pandas as pd
# Read the CSV file into a pandas DataFrame
df = pd.read_csv('DIC_1000.csv')
# Example: Accessing a specific column
DIC_1000= df['DIC']

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 1000 random values
values = DIC_1000
roi41 = 4471
plt.axvline(x=roi41, color='red', linestyle='--', label='ROI41')
# Plot the distribution
plt.hist(values, bins=30, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of 1000 Values')
plt.show()
```
